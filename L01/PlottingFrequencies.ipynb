```python
from collections import Counter


def sliding_window_frequency(S, window_size):
    S = S.upper()
    freq_dict = Counter()
    results = []
    
    for i in range(window_size):
        if S[i].isalpha():
            freq_dict[S[i]] += 1
    results.append(freq_dict.copy())
    
    for i in range(window_size, len(S)):
        if S[i - window_size].isalpha():
            freq_dict[S[i - window_size]] -= 1
            
            if freq_dict[S[i - window_size]] == 0:
                del freq_dict[S[i - window_size]]
                
            if S[i].isalpha():
                freq_dict[S[i]] += 1
            results.append(freq_dict.copy())
            
    return results
```

```python
from matplotlib import pyplot as plt


def plot_sliding_window_frequencies(S,  window_size):
    
    results = sliding_window_frequency(S, window_size)
    
    plt.figure(figsize = (10, 6))
    
    for char in set(S.upper()):
        if char.isalpha():
            char_freqs = [result.get(char, 0) for result in results]
            plt.plot(char_freqs, label = char)
            
    plt.xlabel('Window Position')
    plt.ylabel('Frequency')
    plt.legend(title = "Characters")
    plt.show()
```

```python
S = "cascascascascdacascaacsaccabcasbcasbcasbabbacssabcbcascadadsaddasdbdsabcabcasbcsabcasbcasbcsabcsabacsbcsabca"
window_size = 30

plot_sliding_window_frequencies(S, window_size)
```
